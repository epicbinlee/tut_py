```python
# pip install opencv-python==4.5.3.56
import cv2
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib inline
```

```python
# cat = cv2.imread('pic.png', cv2.IMREAD_GRAYSCALE)
cat = cv2.imread('pic.png')
_, thresh1 = cv2.threshold(cat, 127, 255, cv2.THRESH_BINARY)
_, thresh2 = cv2.threshold(cat, 127, 255, cv2.THRESH_BINARY_INV)
_, thresh3 = cv2.threshold(cat, 127, 255, cv2.THRESH_TRUNC)
_, thresh4 = cv2.threshold(cat, 127, 255, cv2.THRESH_TOZERO)
_, thresh5 = cv2.threshold(cat, 127, 255, cv2.THRESH_TOZERO_INV)
titles = ['CAT', 'BIN', 'BIN_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images = [cat, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2, 3, i + 1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```

```python
from scipy.interpolate import BSpline
import matplotlib.pyplot as plt

# 给定的控制点、拟合点、节点参数和权重
control_points = np.array([[0, 0], [1, 2], [2, 3], [3, 1], [4, 0]])
fit_points = np.array([[0, 0], [1, 1], [2, 2], [3, 1], [4, 0]])
knots = [0, 0, 0, 1, 2, 3, 3, 3]  # 这是一个示例节点参数，可以根据实际情况调整
weights = [1, 2, 1, 1, 1]

# B-spline 曲线的度数
degree_of_bspline = 3

# 创建 B-spline 曲线对象
bspline = BSpline(knots, control_points, degree_of_bspline)

# 计算曲线上的点
num_points = 1000  # 可根据需要调整点的数量
u = np.linspace(0, 3, num_points)  # 参数值范围通常与节点参数范围相匹配

curve_points = np.array([bspline(u_val) for u_val in u])

# 绘制 B-spline 曲线
plt.figure(figsize=(8, 6))
plt.plot(control_points[:, 0], control_points[:, 1], 'ro-', label='Control Points')
plt.plot(fit_points[:, 0], fit_points[:, 1], 'bo-', label='Fit Points')
plt.plot(curve_points[:, 0], curve_points[:, 1], 'g-', label='B-Spline Curve')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('B-Spline Curve')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.interpolate import BSpline
import matplotlib.pyplot as plt

# 定义节点参数、控制点和度数
knots = [0, 0, 0, 1, 2, 3, 3, 3]  # 节点参数
control_points = np.array([[0, 0], [1, 1], [2, 2], [3, 1], [4, 0]])  # 控制点
degree = 3  # 曲线的度数

# 创建 B-spline 曲线对象
bspline = BSpline(knots, control_points, degree)

# 计算曲线上的点
u_values = np.linspace(0, 3, 1000)  # 参数值范围
curve_points = bspline(u_values)

# 绘制 B-spline 曲线
plt.figure(figsize=(8, 6))
plt.plot(control_points[:, 0], control_points[:, 1], 'ro-', label='Control Points')
plt.plot(curve_points[:, 0], curve_points[:, 1], 'g-', label='B-Spline Curve')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('B-Spline Curve')
plt.grid(True)
plt.show()
```
